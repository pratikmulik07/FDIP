```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("lena_color_256.tif")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Show original image
axes[0].imshow(img_rgb)
axes[0].set_title("Original Image")
axes[0].axis('off')

# Plot RGB histogram
colors = ('r', 'g', 'b')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    axes[1].plot(hist, color=col)
    axes[1].set_xlim([0, 256])

axes[1].set_title("Histogram of Image")
axes[1].set_xlabel("Pixel Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Function to add Salt & Pepper noise
def salt_pepper_noise(image, salt_prob=0.01, pepper_prob=0.01):
    noisy = np.copy(image)
    # Salt noise
    num_salt = np.ceil(salt_prob * image.size)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy[coords[0], coords[1]] = 255
    
    # Pepper noise
    num_pepper = np.ceil(pepper_prob * image.size)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy[coords[0], coords[1]] = 0
    return noisy

# Load grayscale image
img = cv2.imread("lena_color_256.tif")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Salt & Pepper Noise
sp_noisy = salt_pepper_noise(gray, 0.02, 0.02)

# Display image and histogram side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Show noisy image
axes[0].imshow(sp_noisy, cmap='gray')
axes[0].set_title("Salt & Pepper Noise Image")
axes[0].axis('off')

# Plot histogram
axes[1].hist(sp_noisy.ravel(), bins=256, range=[0,256], color='red')
axes[1].set_title("Histogram of Salt & Pepper Noise")
axes[1].set_xlabel("Pixel Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Function to add Gaussian noise
def gaussian_noise(image, mean=0, sigma=20):
    gauss = np.random.normal(mean, sigma, image.shape).reshape(image.shape)
    noisy = image + gauss
    noisy = np.clip(noisy, 0, 255)  # Clip to valid range
    return noisy.astype(np.uint8)

# Load grayscale image
img = cv2.imread("lena_color_256.tif")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Noise
gaussian_noisy = gaussian_noise(gray)

# Display image and histogram side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Show Gaussian noisy image
axes[0].imshow(gaussian_noisy, cmap='gray')
axes[0].set_title("Gaussian Noise Image")
axes[0].axis('off')

# Plot histogram
axes[1].hist(gaussian_noisy.ravel(), bins=256, range=[0,256], color='red')
axes[1].set_title("Histogram of Gaussian Noise")
axes[1].set_xlabel("Pixel Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Function to add Impulse noise
def impulse_noise(image, prob=0.02):
    noisy = np.copy(image)
    num_pixels = int(prob * image.size)
    coords = [np.random.randint(0, i - 1, num_pixels) for i in image.shape]
    noisy[coords[0], coords[1]] = np.random.randint(0, 256, num_pixels)
    return noisy

# Load grayscale image
img = cv2.imread("lena_color_256.tif")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Impulse Noise
impulse_noisy = impulse_noise(gray, 0.02)

# Display image and histogram side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Show Impulse noisy image
axes[0].imshow(impulse_noisy, cmap='gray')
axes[0].set_title("Impulse Noise Image")
axes[0].axis('off')

# Plot histogram
axes[1].hist(impulse_noisy.ravel(), bins=256, range=[0,256], color='red')
axes[1].set_title("Histogram of Impulse Noise")
axes[1].set_xlabel("Pixel Intensity")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
